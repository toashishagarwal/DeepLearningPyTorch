```python
import numpy as np
import pandas as pd
import torch

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/gscdit/Breast-Cancer-Detection/master/data.csv')
df.head()
```

```python
df.shape
```

```python
df.drop(columns=['id','Unnamed: 32'], inplace=True)
```

```python
df.head()
```

# Split the data into TRAINING set & TEST set

```python
X_train, X_test, Y_train, Y_test=train_test_split(df.iloc[:,1:],df.iloc[:,0],test_size=0.2)
```

# Scaling the data because some of the features are in different scales

```python
scaler = StandardScaler()
X_train=scaler.fit_transform(X_train)
X_test=scaler.transform(X_test)
```

```python
X_train
```

```python
Y_train
```

# Since Y train object is in alphabets, our NN wont understand. Hence we have to use Label encoder

```python
encoder = LabelEncoder()
Y_train = encoder.fit_transform(Y_train)
Y_test = encoder.transform(Y_test)
```

```python
Y_train
```

# For our NN, we need tensors. Hence we will convert numpy arrays to tensors

```python
X_train_tensor=torch.from_numpy(X_train)
X_test_tensor=torch.from_numpy(X_test)
Y_train_tensor=torch.from_numpy(Y_train)
Y_test_tensor=torch.from_numpy(Y_test)
```

```python
X_train_tensor
```

# Create a model

```python
class MySimpleModel:
    def __init__(self, X):
      self.weights = torch.rand(X.shape[1],1,dtype=torch.float64,requires_grad=True)
      self.bias = torch.zeros(1,dtype=torch.float64,requires_grad=True)

    def forward(self, X):
      z=torch.matmul(X, self.weights) + self.bias  # z = wx + b
      y_pred = torch.sigmoid(z)                    # y_pred = sigmoid(z)
      return y_pred

    def loss_function(self, y_pred, y):
      epsilon=1e-7
      y_pred=torch.clamp(y_pred,epsilon,1-epsilon)

      # Calculate loss
      loss = -(Y_train_tensor * torch.log(y_pred) + (1 - Y_train_tensor) * torch.log(1 - y_pred)).mean()
      return loss
```

```python
learning_rate=0.1
epochs = 25
```

```python
# Now we need to train the model
# a) Create the model instance
model=MySimpleModel(X_train_tensor)

for epoch in range(epochs):
  # b) Forward pass
  y_pred=model.forward(X_train_tensor)

  # c) loss calculation
  loss=model.loss_function(y_pred,Y_train_tensor)

  # d) back propagation
  loss.backward()

  # e) parameters update
  with torch.no_grad():
    model.weights -= learning_rate * model.weights.grad
    model.bias -= learning_rate * model.bias.grad

  # f) zero the gradients
  model.weights.grad.zero_()
  model.bias.grad.zero_()

  print(f'Epoch: {epoch+1}, Loss: {loss.item()}')
```

# Evaluation

```python
from itertools import accumulate
with torch.no_grad():
  y_pred=model.forward(X_test_tensor)
  y_pred = (y_pred > 0.5).float()
  accuracy = (y_pred == Y_test_tensor).float().mean()
  print(f'Accuracy: {accuracy.item()}')
```
